```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()


seed = 0
N = 100
M = 3
b = 1
a = 0.001
np.random.seed(seed)


def sample_paths(M, N, b, a):
    dt = 1.0 / N
    B = np.empty((M, N), dtype=np.float32)
    B[:, 0] = a                                   # X(0) = a
    for n in range(N - 1):
        t = n * dt
        xi = np.random.randn(M) * np.sqrt(dt)
        B[:, n + 1] = B[:, n] * (b - dt / (1 - t)) + xi
        #B[:, n + 1] = B[:, n] * (a-b)*dt + xi
    return B


B = sample_paths(M, N, b, a)
plt.title("Three Sample paths")
plt.plot(B.T)
#plt.show()
plt.savefig("brownian_bridge.png")
```
